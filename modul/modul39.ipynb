#### Фильтрация с помощью str.match и str.contains
Очень часто приходится фильтровать значения в столбце по соответствию строке или её части.
Для этого есть две полезные функции:

1. str.match("abc") ищет строки которые НАЧИНАЮТСЯ c "abc";
2. str.contains("abc") ищет строки в которых ЕСТЬ "abc".
3. ~str.match("abc") ищет строки которые НЕ НАЧИНАЮТСЯ c "abc";
4. ~str.contains("abc") ищет строки в которых НЕТ "abc".

Обе функции не могут работать с NaN. По этой причине необходимо использовать параметр na = False.

```python
sample.name.str.match("К", na=False)
```

```python
sample[sample.name.str.match("К", na=False)]
```

```python
sample[~sample.name.str.match("К", na=False)]
```

## 39.7 Преобразование данных

#### метод apply

df.column_name.apply(func)

Функции, с которыми работает apply, бывают двух видов:

1. Лямбда-функции. Их вы можете задать прямо внутри apply.
2. Обычные именные функции. Их вы определяете через def.

```python
"""Возведем возраст в sample в квадрат:"""

sample.age.apply(lambda x:x**2)
```

```python
"""Обнулим значения возраста для всех возрастов больше 19:"""

def func(x):
    if x < 20:
        return x
    else:
        return 0

sample.age.apply(func)
```

```python
"""С помощью apply и lambda-функции увеличьте возраст во всех записях на 1 год и сохраните
в sample2. Необходимо присвоить весь исправленный датафрейм."""
sample2 = sample
sample2.age = sample.age.apply(lambda x:x+1)
sample2
```

```python
"""С помощью apply и lambda-функции замените все буквы в поле City на строчные и сохраните 
в sample2. Примечание. Вам может понадобится функция s.lower(). Когда в столбце есть 
пропущенные значения, необходимо в явном виде указывать, что это str: str(s).lower()."""

sample2 = sample
sample2.city = sample.city.apply(lambda x:str(x).lower())
sample2
```
